# Project Background

**Introduction:**

Music has been a source of comfort and inspiration for humans for centuries. In recent times, it has gained popularity as a form of therapy for mental health issues. The purpose of this study was to assess the impact of music on people's mental health and mood in general.

**Methodology:**

A survey was conducted to gather information from participants about their musical preferences, listening habits, and overall state of mind. The survey was completed by over 700 participants and recorded their responses. Respondents were given 16 genres of music and asked to rank the frequency of listening to each, with options ranging from never to very frequently. They were also asked to rank the intensity of their feelings of anxiety, depression, insomnia, and OCD on a scale of 1 to 10, with 0 being the lowest in intensity and 10 being the highest.

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the survey data into a Pandas dataframe
df = pd.read_csv("/content/Music-and-Therapy-survey-1.csv")

# Check the column names in the dataframe
print(df.columns)

 
```

```python
# Plot the distribution of age using a histogram
plt.hist(df["Age"], bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Age")
```

```python
# Plot the frequency of music listening for each genre using a bar graph
sns.countplot(x="Frequency [Classical]", hue="Anxiety", data=df)
plt.xticks(rotation=200)
plt.xlabel("Frequency [Classical]")
plt.ylabel("Frequency")
plt.title("Frequency of Music Listening for each Genre")
```

```python
# Plot the frequency of music listening for each genre using a bar graph
sns.countplot(x="Frequency [Classical]", hue="Age", data=df)
plt.xticks(rotation=20)
plt.xlabel("Frequency [Classical]")
plt.ylabel("Frequency")
plt.title("Frequency of Music Listening for each Genre")
```

```python
# Plot the relationship between anxiety levels and hours per day listening to music
sns.scatterplot(x="Hours per day", y="Anxiety", data=df)
plt.xlabel("Hours per day")
plt.ylabel("Anxiety")
plt.title("Relationship between Hours per day and Anxiety")
```

```python
# Plot the correlation between the intensity of mental health issues and the frequency of listening to music
sns.jointplot(x=df["Anxiety"], y=df["Frequency [Classical]"], kind="reg")
plt.xlabel("Anxiety")
plt.ylabel("Frequency of listening [Classical]")
plt.title("Correlation between anxiety and frequency of listening to classical music")
```

```python
# Bin the age data into different age groups
bins = [0, 18, 25, 35, 45, 60, 100]
df["Age group"] = pd.cut(df["Age"], bins, labels=["<18", "18-24", "25-34", "35-44", "45-59", "60+"])

# Plot the frequency of listening for each genre by age group using a violin plot
sns.plot(x=df["Age"], y=df["Frequency [Pop]"])
plt.xlabel("Age group")
plt.ylabel("Frequency of listening [Pop]")
plt.title("Frequency of listening for Pop genre by age group")
```

```python
# Group the data by the primary streaming service and count the occurrences
grouped = df.groupby("Primary streaming service").size().reset_index(name='counts')

# Plot the data as a pie chart
plt.pie(grouped["counts"], labels=grouped["Primary streaming service"], startangle=90, autopct='%1.1f%%')
plt.axis('equal')  # Make the pie chart a circle
plt.show()
```
